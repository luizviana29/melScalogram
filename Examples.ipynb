```python
from melscalogram import MelScalogram
import librosa
```

```python
signal, sr = librosa.load(librosa.ex('brahms'))
mel_scalogram = MelScalogram(signal, sr, wavelet='cmor1-1', n_scales=40, lower_hz=20, upper_hz=8000)
```

```python
mel_scalogram.plt_mean_mel_scalogram()
```

```python
mel_scalogram.plt_uniform_mel_scalogram()
```

```python
mel_scalogram.plt_max_mel_scalogram()
```

```python
mel_scalogram.plt_min_mel_scalogram()
```

```python
mel_scalogram.mean_for_cnn(x_pixels=512, y_pixels=40, fname='mean_mel_scalogram_for_cnn.jpg')
mel_scalogram.uniform_for_cnn(x_pixels=512, y_pixels=40, fname='uniform_mel_scalogram_for_cnn.jpg')
mel_scalogram.max_for_cnn(x_pixels=512, y_pixels=40, fname='max_mel_scalogram_for_cnn.jpg')
mel_scalogram.min_for_cnn(x_pixels=512, y_pixels=40, fname='min_mel_scalogram_for_cnn.jpg')
```
